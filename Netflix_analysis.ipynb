# 4번

### description 합친 str

```python
desc = df['description'].sum()
desc = desc.lower()
print(desc)
```

```python
type(desc)
```

### nltk로 stopword 제거

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 

stop_words = set(stopwords.words('english')) 

word_tokens = word_tokenize(desc)

stop_str = ""
for token in word_tokens:
    if token not in stop_words:
        stop_str += (token+" ")

print(stop_str) 
```

```python
type(stop_str)
```

###  문장부호 제거

```python
import re
word_str = stop_str.replace("\'s", "")
word_str = re.sub(r'[!\'"#$%&\()*+,-./:;<=>?@\[\]^_\`{|}~\\\\]','', word_str)

print(word_str)
```

```python
word_list = word_str.split()
print(word_list)
```

```python
word_list_no_duplicate = set(word_list)
word_count = []

for word in word_list_no_duplicate:
    word_count.append((word_list.count(word), word))
    
for result in sorted(word_count, reverse=True):
    n += 1
    print(result[1]+ ':', result[0])
```
